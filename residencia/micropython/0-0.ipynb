```python
import math
# Exemplo: Linearização de NTC (Equação Steinhart-Hart simplificada)
resistencia = 10000.0
beta = 3950.0 # T = 1 / ( (1/T0) + (1/B) * ln(R/R0) )
temp_k = 1 / ( (1/298.15) + (1/beta) * math.log(resistencia / 10000.0) )
print(f"Temperatura: {temp_k - 273.15:.2f}ºC")
# Exemplo: Seno de 90 graus
seno = math.sin(math.radians(90))
print(f"Seno de 90º: {seno}")
```

```python
import time
import random
print("--- SISTEMA ONLINE (CTRL+C para parar) ---")
while True:
  # 1. Leitura
  temp = random.uniform(20.0, 35.0)
  # 2. Decisão
  if temp > 30.0:
    status = "ALERTA"
  else:
    status = "Normal"
  # 3. Saída
  print(f"Temp: {temp:.1f} | Status: {status}")
  # 4. Delay (Vital para não travar a CPU)
  time.sleep(2)
  # Trava de segurança para o exemplo não rodar para sempre aqui
  if temp > 34.0:
    break
```

```python
# Função flexível que aceita argumentos
def controlar_motor(numero_motor, velocidade):
  # Validando dados
  if velocidade > 100:
    velocidade = 100
    print("Aviso: Velocidade limitada a 100%")
  print(f"⏩ Motor {numero_motor} ativado com força {velocidade}%")
# Chamando com dados diferentes
controlar_motor(1, 50)
controlar_motor(2, 150) # Vai ativar a proteção
controlar_motor(1, 0)
```

```python
def converter_adc_para_volts(leitura_bruta):
  # O ESP32 tem 12 bits (0 a 4095) e trabalha com 3.3V
  tensao = (leitura_bruta * 3.3) / 4095.0
  return tensao # Devolve o valor calculado
# Simulando leitura do sensor
valor_lido = 2048
# Capturando a resposta da função
volts = converter_adc_para_volts(valor_lido)
print(f"Leitura ADC: {valor_lido}")
print(f"Tensão Real: {volts:.2f}V")
# Podemos usar a função direto numa condição
if converter_adc_para_volts(4000) > 3.0:
  print("Perigo: Tensão alta!")
```
